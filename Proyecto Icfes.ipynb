```python
def contar_outliers_por_estrato(df, col_estrato, col_puntaje):

# Calcula y cuenta los valores atípicos por estrato usando el método del boxplot
    outliers_info = {}
    
    for estrato in sorted(df[col_estrato].unique()):
        datos_estrato = df[df[col_estrato] == estrato][col_puntaje]
        
        # Calcular cuartiles y rango intercuartílico (IQR)
        Q1 = datos_estrato.quantile(0.25)
        Q3 = datos_estrato.quantile(0.75)
        IQR = Q3 - Q1
        
        # Definir límites para valores atípicos
        limite_inferior = Q1 - 1.5 * IQR
        limite_superior = Q3 + 1.5 * IQR
        
        # Contar outliers
        outliers_inferiores = len(datos_estrato[datos_estrato < limite_inferior])
        outliers_superiores = len(datos_estrato[datos_estrato > limite_superior])
        total_outliers = outliers_inferiores + outliers_superiores
        
        outliers_info[estrato] = {
            'total_estrato': len(datos_estrato),
            'outliers_inferiores': outliers_inferiores,
            'outliers_superiores': outliers_superiores,
            'total_outliers': total_outliers,
            'porcentaje_outliers': (total_outliers / len(datos_estrato)) * 100,
            'limite_inferior': limite_inferior,
            'limite_superior': limite_superior,
            'Q1': Q1,
            'Q3': Q3,
            'mediana': datos_estrato.median()
        }
    
    return outliers_info

# Calcular outliers
outliers_por_estrato = contar_outliers_por_estrato(df_clean, 'fami_estratovivienda', 'mod_ingles_punt')

# Mostrar resultados
print("\nANÁLISIS DE VALORES ATÍPICOS POR ESTRATO")

for estrato, info in outliers_por_estrato.items():
    print(f"\nEstrato {estrato}:")
    print(f"  Total registros: {info['total_estrato']:,}")
    print(f"  Outliers inferiores: {info['outliers_inferiores']} ({info['outliers_inferiores']/info['total_estrato']*100:.2f}%)")
    print(f"  Outliers superiores: {info['outliers_superiores']} ({info['outliers_superiores']/info['total_estrato']*100:.2f}%)")
    print(f"  Total outliers: {info['total_outliers']} ({info['porcentaje_outliers']:.2f}%)")
    print(f"  Límite inferior: {info['limite_inferior']:.2f}")
    print(f"  Límite superior: {info['limite_superior']:.2f}")
    print(f"  Q1: {info['Q1']:.2f}, Q3: {info['Q3']:.2f}, Mediana: {info['mediana']:.2f}")

# Visualización con boxplot que incluye información de outliers
plt.figure(figsize=(14, 8))

# Crear boxplot
boxplot = df_clean.boxplot(column='mod_ingles_punt', by='fami_estratovivienda', 
                          grid=True, figsize=(14, 8))
y_min, y_max = plt.ylim()
x_positions = range(1, len(outliers_por_estrato) + 1)
# Personalizar el gráfico
plt.xlabel('Estrato Socioeconómico', fontsize=12, fontweight='bold')
plt.ylabel('Puntaje de Inglés Saber Pro', fontsize=12, fontweight='bold')
plt.title('Distribución de Puntajes de Inglés por Estrato\n(Con análisis de valores atípicos)', 
          fontsize=14, fontweight='bold')
plt.suptitle('')  # Eliminar título automático

# Agregar anotaciones con el conteo de outliers
for estrato, info in outliers_por_estrato.items():
    if info['total_outliers'] > 0:
        # Calcular posición vertical inteligente
        # Usar una posición relativa al rango del gráfico
        y_position = y_max - (y_max - y_min) * 0.09  # 5% desde el top
        plt.text(estrato, y_position, 
                f"Outliers: {info['total_outliers']}\n({info['porcentaje_outliers']:.1f}%)",
                ha='center', va='top', fontsize=10, fontweight='bold',
                bbox=dict(boxstyle="round,pad=0.4", facecolor="yellow", alpha=0.8, edgecolor='black'))
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

# Gráfico adicional: porcentaje de outliers por estrato
plt.figure(figsize=(15, 8))

estratos = list(outliers_por_estrato.keys())
porcentajes = [info['porcentaje_outliers'] for info in outliers_por_estrato.values()]

plt.bar(estratos, porcentajes, color='lightcoral', edgecolor='darkred', linewidth=1)
plt.xlabel('Estrato Socioeconómico', fontsize=12, fontweight='bold')
plt.ylabel('Porcentaje de Valores Atípicos (%)', fontsize=12, fontweight='bold')
plt.title('Porcentaje de Valores Atípicos por Estrato', fontsize=14, fontweight='bold')

# Agregar valores en las barras
for i, porcentaje in enumerate(porcentajes):
    plt.text(estratos[i], porcentaje + 0.1, f'{porcentaje:.1f}%', 
             ha='center', va='bottom', fontweight='bold')

plt.grid(True, alpha=0.3, axis='y')
plt.xticks(estratos)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from scipy.stats import kruskal
import scikit_posthocs as sp
import warnings
warnings.filterwarnings('ignore')

# Cargar los datos
df = pd.read_excel("D:\Anaconda\Saber Pro 20232.xlsx", sheet_name='DATA_ICFES_SABERPRO_PROFESIONAL')

# SOLUCIÓN 1: Convertir la columna de estrato a string y filtrar solo valores numéricos
df['fami_estratovivienda'] = df['fami_estratovivienda'].astype(str)

# Filtrar solo los estratos que contienen números (eliminar "Sin estrato", "nan", etc.)
# Usamos una expresión regular para identificar strings que contienen dígitos
df_numeric_estratos = df[df['fami_estratovivienda'].str.contains(r'\d', na=False)].copy()

# Convertir los estratos a numérico para mejor ordenamiento
df_numeric_estratos['fami_estratovivienda'] = pd.to_numeric(df_numeric_estratos['fami_estratovivienda'], errors='coerce')

# Eliminar posibles NaN que pudieron generarse en la conversión
df_numeric_estratos = df_numeric_estratos.dropna(subset=['fami_estratovivienda'])

# Verificar las primeras filas y estructura de los datos
print("Primeras 5 filas del dataset filtrado:")
print(df_numeric_estratos[['fami_estratovivienda', 'mod_ingles_punt']].head())

# Verificar valores únicos en estrato después del filtrado
print("\nValores únicos en estrato después del filtrado:", sorted(df_numeric_estratos['fami_estratovivienda'].unique()))

# Limpiar datos: eliminar filas con valores nulos en las columnas de interés
df_clean = df_numeric_estratos[['fami_estratovivienda', 'mod_ingles_punt']].dropna()

print(f"\nDatos originales: {len(df)} filas")
print(f"Datos después de filtrar estratos no numéricos: {len(df_numeric_estratos)} filas")
print(f"Datos limpios para análisis: {len(df_clean)} filas")

# Separar los puntajes por estrato
estratos = sorted(df_clean['fami_estratovivienda'].unique())
puntajes_por_estrato = []

print("\n" + "="*50)
print("ESTADÍSTICOS DESCRIPTIVOS POR ESTRATO (SOLO NUMÉRICOS)")
print("="*50)

for estrato in estratos:
    puntajes = df_clean[df_clean['fami_estratovivienda'] == estrato]['mod_ingles_punt']
    puntajes_por_estrato.append(puntajes)
    
    print(f"Estrato {int(estrato)}:")
    print(f"  Número de estudiantes: {len(puntajes):,}")
    print(f"  Media: {puntajes.mean():.2f}")
    print(f"  Desviación estándar: {puntajes.std():.2f}")
    print(f"  Mediana: {puntajes.median():.2f}")
    print("-" * 30)

# PRUEBA DE NORMALIDAD (Shapiro-Wilk) para cada estrato
print("\n" + "="*50)
print("PRUEBAS DE NORMALIDAD - SHAPIRO-WILK")
print("="*50)

for i, estrato in enumerate(estratos):
    # Tomar una muestra más pequeña para Shapiro-Wilk (máximo 5000)
    muestra = puntajes_por_estrato[i].sample(n=min(5000, len(puntajes_por_estrato[i])), random_state=42)
    stat, p_value = stats.shapiro(muestra)
    print(f"Estrato {int(estrato)}: Estadístico = {stat:.4f}, p-value = {p_value:.4f}")

# PRUEBA DE HOMOCEDASTICIDAD (Levene)
print("\n" + "="*50)
print("PRUEBA DE HOMOCEDASTICIDAD - LEVENE")
print("="*50)

stat_levene, p_levene = stats.levene(*puntajes_por_estrato)
print(f"Estadístico de Levene: {stat_levene:.4f}")
print(f"p-value: {p_levene:.4f}")

# PRUEBA DE KRUSKAL-WALLIS
print("\n" + "="*50)
print("PRUEBA DE KRUSKAL-WALLIS")
print("="*50)

stat_kw, p_kw = kruskal(*puntajes_por_estrato)

print(f"Estadístico de Kruskal-Wallis: H = {stat_kw:.4f}")
print(f"p-value: {p_kw:.4f}")

if p_kw < 0.05:
    print("→ Resultado: Existen diferencias estadísticamente significativas entre al menos dos estratos")
    print("→ Se rechaza la hipótesis nula")
else:
    print("→ Resultado: No existen diferencias estadísticamente significativas entre los estratos")
    print("→ No se rechaza la hipótesis nula")

# PRUEBA POST-HOC DE DUNN
print("\n" + "="*50)
print("PRUEBA POST-HOC DE DUNN")
print("="*50)

# Preparar datos para Dunn
dunn_data = []
estrato_labels = []
for estrato, puntajes in zip(estratos, puntajes_por_estrato):
    dunn_data.extend(puntajes)
    estrato_labels.extend([f'Estrato {int(estrato)}'] * len(puntajes))

dunn_df = pd.DataFrame({
    'puntaje': dunn_data,
    'estrato': estrato_labels
})

# Realizar prueba de Dunn
dunn_result = sp.posthoc_dunn(dunn_df, val_col='puntaje', group_col='estrato', p_adjust='bonferroni')

# MODIFICACIÓN: Formatear la matriz para mostrar solo 2 decimales
print("Matriz de p-values ajustados (Dunn):")
print("(Valores mostrados con 2 decimales)")

# Crear una copia de la matriz formateada con 2 decimales
dunn_formatted = dunn_result.copy()

# Aplicar formato de 2 decimales a todos los valores
for col in dunn_formatted.columns:
    dunn_formatted[col] = dunn_formatted[col].apply(lambda x: f'{x:.2f}')

print(dunn_formatted)

# También mostrar la matriz original pero con formato mejorado
print("\nMatriz de p-values (valores numéricos completos):")
print(dunn_result.round(4))

# Interpretar resultados de Dunn
print("\n" + "="*50)
print("COMPARACIONES SIGNIFICATIVAS (p < 0.05)")
print("="*50)

significativas = []
no_significativas = []

for i in range(len(dunn_result)):
    for j in range(i+1, len(dunn_result)):
        p_val = dunn_result.iloc[i,j]
        if p_val < 0.05:
            significativas.append((dunn_result.index[i], dunn_result.columns[j], p_val))
        else:
            no_significativas.append((dunn_result.index[i], dunn_result.columns[j], p_val))

if significativas:
    print("COMPARACIONES ESTADÍSTICAMENTE SIGNIFICATIVAS:")
    for comparacion in significativas:
        print(f"  ✓ {comparacion[0]} vs {comparacion[1]}: p = {comparacion[2]:.4f} ***")
else:
    print("No se encontraron comparaciones significativas")

if no_significativas:
    print(f"\nCOMPARACIONES NO SIGNIFICATIVAS ({len(no_significativas)} pares):")
    # Mostrar solo algunas comparaciones no significativas para no saturar
    if len(no_significativas) > 10:
        print("  (Mostrando 10 de {len(no_significativas)} comparaciones no significativas)")
        for comparacion in no_significativas[:10]:
            print(f"  • {comparacion[0]} vs {comparacion[1]}: p = {comparacion[2]:.4f}")
    else:
        for comparacion in no_significativas:
            print(f"  • {comparacion[0]} vs {comparacion[1]}: p = {comparacion[2]:.4f}")

# VISUALIZACIONES
print("\n" + "="*50)
print("GENERANDO GRÁFICOS")
print("="*50)

# Crear figura con subplots
fig, axes = plt.subplots(2, 2, figsize=(15, 12))

# Gráfico 1: Boxplot por estrato
data_for_plot = []
labels = []
for estrato, puntajes in zip(estratos, puntajes_por_estrato):
    data_for_plot.append(puntajes)
    labels.append(f'Estrato {int(estrato)}\n(n={len(puntajes):,})')

axes[0,0].boxplot(data_for_plot, labels=labels)
axes[0,0].set_title('Distribución de Puntajes por Estrato Socioeconómico\n(Solo estratos numéricos)', fontsize=14, fontweight='bold')
axes[0,0].set_ylabel('Puntaje de Inglés')
axes[0,0].grid(True, alpha=0.3)
axes[0,0].tick_params(axis='x', rotation=45)

# Gráfico 2: Medias por estrato
medias = [puntajes.mean() for puntajes in puntajes_por_estrato]
desviaciones = [puntajes.std() for puntajes in puntajes_por_estrato]

axes[0,1].bar(labels, medias, yerr=desviaciones, capsize=5, alpha=0.7, color='skyblue')
axes[0,1].set_title('Media de Puntajes por Estrato con Desviación Estándar\n(Solo estratos numéricos)', fontsize=14, fontweight='bold')
axes[0,1].set_ylabel('Puntaje Promedio')
axes[0,1].grid(True, alpha=0.3)
axes[0,1].tick_params(axis='x', rotation=45)

# Añadir valores en las barras
for i, v in enumerate(medias):
    axes[0,1].text(i, v + desviaciones[i] + 1, f'{v:.1f}', ha='center', va='bottom', fontweight='bold')

# Gráfico 3: Heatmap de la matriz de Dunn
# Crear máscara para mostrar solo la triangular superior (sin diagonal)
mask = np.triu(np.ones_like(dunn_result, dtype=bool), k=1)

sns.heatmap(dunn_result, 
            mask=mask,
            annot=True, 
            fmt='.3f', 
            cmap='RdYlBu_r', 
            center=0.05,
            cbar_kws={'label': 'p-value'},
            ax=axes[1,0])
axes[1,0].set_title('Matriz de p-values de Dunn\n(Prueba Post-Hoc)', fontsize=14, fontweight='bold')
axes[1,0].tick_params(axis='x', rotation=45)
axes[1,0].tick_params(axis='y', rotation=0)

# Gráfico 4: Tendencia de medias
axes[1,1].plot(range(len(estratos)), medias, marker='o', linewidth=2, markersize=8, color='red')
axes[1,1].set_xticks(range(len(estratos)))
axes[1,1].set_xticklabels([f'Estrato {int(e)}' for e in estratos])
axes[1,1].fill_between(range(len(estratos)), 
                      [m - d for m, d in zip(medias, desviaciones)], 
                      [m + d for m, d in zip(medias, desviaciones)], 
                      alpha=0.2, color='red')
axes[1,1].set_title('Tendencia del Puntaje Promedio por Estrato\n(Solo estratos numéricos)', fontsize=14, fontweight='bold')
axes[1,1].set_xlabel('Estrato Socioeconómico')
axes[1,1].set_ylabel('Puntaje Promedio')
axes[1,1].grid(True, alpha=0.3)

# Añadir valores en la línea de tendencia
for i, (x, y) in enumerate(zip(range(len(estratos)), medias)):
    axes[1,1].annotate(f'{y:.1f}', (x, y), textcoords="offset points", xytext=(0,10), ha='center')

plt.tight_layout()
plt.show()

# RESUMEN EJECUTIVO
print("\n" + "="*70)
print("RESUMEN EJECUTIVO DEL ANÁLISIS")
print("="*70)

print(f"✓ Total de estudiantes analizados: {len(df_clean):,}")
print(f"✓ Número de estratos analizados: {len(estratos)}")
print(f"✓ Estratos incluidos: {[int(e) for e in estratos]}")
print(f"✓ Prueba de Kruskal-Wallis: H = {stat_kw:.4f}, p = {p_kw:.4f}")

if p_kw < 0.05:
    print("✓ CONCLUSIÓN: Existe evidencia estadística para afirmar que hay diferencias")
    print("  significativas en los puntajes entre al menos dos estratos socioeconómicos.")
    print(f"✓ Número de comparaciones significativas encontradas: {len(significativas)}")
    
    # Resumen de diferencias significativas por estrato
    estratos_con_diferencias = set()
    for comp in significativas:
        estratos_con_diferencias.add(comp[0])
        estratos_con_diferencias.add(comp[1])
    
    print(f"✓ Estratos con diferencias significativas: {sorted(estratos_con_diferencias)}")
else:
    print("✓ CONCLUSIÓN: No existe evidencia estadística suficiente para afirmar")
    print("  que hay diferencias significativas entre los estratos socioeconómicos.")

print(f"✓ Estrato con mayor puntaje promedio: {int(estratos[np.argmax(medias)])} ({max(medias):.1f})")
print(f"✓ Estrato con menor puntaje promedio: {int(estratos[np.argmin(medias)])} ({min(medias):.1f})")
print(f"✓ Diferencia entre mayor y menor estrato: {max(medias) - min(medias):.1f} puntos")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from scipy.stats import kruskal
import scikit_posthocs as sp
import pingouin as pg
import warnings
warnings.filterwarnings('ignore')

# Cargar y preparar los datos
df = pd.read_excel("D:\Anaconda\Saber Pro 20232.xlsx", sheet_name='DATA_ICFES_SABERPRO_PROFESIONAL')

# Convertir y filtrar estratos numéricos
df['fami_estratovivienda'] = df['fami_estratovivienda'].astype(str)
df_numeric_estratos = df[df['fami_estratovivienda'].str.contains(r'\d', na=False)].copy()
df_numeric_estratos['fami_estratovivienda'] = pd.to_numeric(df_numeric_estratos['fami_estratovivienda'], errors='coerce')
df_numeric_estratos = df_numeric_estratos.dropna(subset=['fami_estratovivienda'])
df_clean = df_numeric_estratos[['fami_estratovivienda', 'mod_ingles_punt']].dropna()

# Separar puntajes por estrato
estratos = sorted(df_clean['fami_estratovivienda'].unique())
puntajes_por_estrato = [df_clean[df_clean['fami_estratovivienda'] == estrato]['mod_ingles_punt'] for estrato in estratos]

# =============================================================================
# FUNCIÓN CORREGIDA PARA CLIFF'S DELTA
# =============================================================================

def cliffs_delta(x, y):
    """
    Calcula Cliff's Delta para dos grupos de datos
    """
    x = np.array(x)
    y = np.array(y)
    
    # Matriz de dominancia
    dominance_matrix = np.where(x[:, np.newaxis] > y, 1, 
                               np.where(x[:, np.newaxis] < y, -1, 0))
    
    # Cálculo de Cliff's Delta
    delta = np.mean(dominance_matrix)
    
    return delta

print("═" * 70)
print("ANÁLISIS ESTADÍSTICO COMPLETO - PRUEBAS SABER PRO INGLÉS")
print("═" * 70)

# =============================================================================
# 1. CÁLCULO DE ESTADÍSTICOS DESCRIPTIVOS POR ESTRATO
# =============================================================================

def calcular_ic_95_mediana(data):
    """Calcula el intervalo de confianza 95% para la mediana usando bootstrap"""
    n_bootstraps = 1000
    medians = []
    for _ in range(n_bootstraps):
        sample = np.random.choice(data, size=len(data), replace=True)
        medians.append(np.median(sample))
    return np.percentile(medians, [2.5, 97.5])

print("\n1. ESTADÍSTICOS DESCRIPTIVOS POR ESTRATO")
print("-" * 50)

estadisticos_por_estrato = []

for i, estrato in enumerate(estratos):
    puntajes = puntajes_por_estrato[i]
    media = puntajes.mean()
    mediana = puntajes.median()
    desviacion = puntajes.std()
    ic_mediana = calcular_ic_95_mediana(puntajes)
    n = len(puntajes)
    
    estadisticos_por_estrato.append({
        'Estrato': int(estrato),
        'n': n,
        'Media': media,
        'Mediana': mediana,
        'DE': desviacion,
        'IC95%_inf': ic_mediana[0],
        'IC95%_sup': ic_mediana[1]
    })
    
    print(f"Estrato {int(estrato)}:")
    print(f"  n: {n:,}")
    print(f"  Media: {media:.2f}")
    print(f"  Mediana: {mediana:.2f}")
    print(f"  DE: {desviacion:.2f}")
    print(f"  IC95% Mediana: [{ic_mediana[0]:.2f}, {ic_mediana[1]:.2f}]")
    print("-" * 30)

# Crear DataFrame con estadísticos
df_estadisticos = pd.DataFrame(estadisticos_por_estrato)

# =============================================================================
# 2. PRUEBA DE KRUSKAL-WALLIS Y TAMAÑO DEL EFECTO
# =============================================================================

print("\n2. PRUEBA DE KRUSKAL-WALLIS Y TAMAÑO DEL EFECTO")
print("-" * 50)

# Realizar prueba de Kruskal-Wallis
stat_kw, p_kw = kruskal(*puntajes_por_estrato)
print(f"Kruskal-Wallis: H = {stat_kw:.4f}, p = {p_kw:.4f}")

# Calcular tamaño del efecto (epsilon-squared)
def calcular_epsilon_squared(h_stat, n_total, k_groups):
    """Calcula epsilon-squared como tamaño del efecto para Kruskal-Wallis"""
    return (h_stat - k_groups + 1) / (n_total - k_groups)

n_total = len(df_clean)
k_groups = len(estratos)
epsilon_sq = calcular_epsilon_squared(stat_kw, n_total, k_groups)

# Interpretación del tamaño del efecto
if epsilon_sq < 0.06:
    magnitud = "efecto pequeño"
elif epsilon_sq < 0.14:
    magnitud = "efecto moderado"
else:
    magnitud = "efecto grande"

print(f"Tamaño del efecto (epsilon-squared): {epsilon_sq:.4f} ({magnitud})")

# =============================================================================
# 3. PRUEBA POST-HOC DE DUNN CON P-VALUES AJUSTADOS
# =============================================================================

print("\n3. PRUEBA POST-HOC DE DUNN")
print("-" * 50)

# Preparar datos para Dunn
dunn_data = []
estrato_labels = []
for estrato, puntajes in zip(estratos, puntajes_por_estrato):
    dunn_data.extend(puntajes)
    estrato_labels.extend([f'Estrato {int(estrato)}'] * len(puntajes))

dunn_df = pd.DataFrame({
    'puntaje': dunn_data,
    'estrato': estrato_labels
})

# Realizar prueba de Dunn con ajuste Bonferroni
dunn_result = sp.posthoc_dunn(dunn_df, val_col='puntaje', group_col='estrato', p_adjust='bonferroni')

print("Matriz de p-values ajustados (Dunn):")
dunn_formatted = dunn_result.copy()
for col in dunn_formatted.columns:
    dunn_formatted[col] = dunn_formatted[col].apply(lambda x: f'{x:.4f}')
print(dunn_formatted)

# =============================================================================
# 4. TAMAÑOS DEL EFECTO POR PARES (CLIFF'S DELTA)
# =============================================================================

# Calcular Cliff's Delta para cada par de estratos usando la función optimizada
print("Cliff's Delta para comparaciones por pares:")
print("(Valores > 0 indican que el grupo de la fila tiene puntajes más altos)")

effect_sizes = []

for i in range(len(estratos)):
    for j in range(i+1, len(estratos)):
        grupo_a = puntajes_por_estrato[i]
        grupo_b = puntajes_por_estrato[j]
        
        # Calcular Cliff's Delta usando la función optimizada
        delta = cliffs_delta(grupo_a, grupo_b)
        
        # Interpretación de Cliff's Delta
        if abs(delta) < 0.11:
            interpretacion = "efecto pequeño"
        elif abs(delta) < 0.28:
            interpretacion = "efecto medio"
        elif abs(delta) < 0.43:
            interpretacion = "efecto grande"
        else:
            interpretacion = "efecto muy grande"
        
        effect_sizes.append({
            'Grupo_A': f'Estrato {int(estratos[i])}',
            'Grupo_B': f'Estrato {int(estratos[j])}',
            'Cliffs_Delta': delta,
            'Interpretacion': interpretacion,
            'Magnitud_absoluta': interpretacion
        })
        
        print(f"Estrato {int(estratos[i])} vs Estrato {int(estratos[j])}: δ = {delta:.4f} ({interpretacion})")

df_effect_sizes = pd.DataFrame(effect_sizes)

# =============================================================================
# 5. TABLA RESUMEN COMPLETA SOLICITADA POR EL PROFESOR
# =============================================================================

print("\n5. TABLA RESUMEN COMPLETA")
print("═" * 70)

# Crear tabla resumen consolidada
tabla_resumen = df_estadisticos.copy()

# Añadir información de comparaciones significativas
comparaciones_significativas = []
for i, estrato_a in enumerate(estratos):
    signif = []
    for j, estrato_b in enumerate(estratos):
        if i != j:
            p_val = dunn_result.iloc[i, j]
            if p_val < 0.05:
                signif.append(f"E{int(estrato_b)}")
    comparaciones_significativas.append(", ".join(signif) if signif else "Ninguna")

tabla_resumen['Dif_significativas'] = comparaciones_significativas

# Mostrar tabla resumen completa
print("\nTABLA 1: RESUMEN ESTADÍSTICO POR ESTRATO SOCIOECONÓMICO")
print("=" * 80)
print(f"{'Estrato':<8} {'n':<12} {'Media':<8} {'Mediana':<10} {'DE':<8} {'IC95% Mediana':<20} {'Dif. Signif.'}")
print("-" * 80)

for _, row in tabla_resumen.iterrows():
    print(f"{int(row['Estrato']):<8} {row['n']:<12,} {row['Media']:<8.2f} {row['Mediana']:<10.2f} {row['DE']:<8.2f} [{row['IC95%_inf']:.2f}, {row['IC95%_sup']:.2f}]  {row['Dif_significativas']}")

# =============================================================================
# 6. VISUALIZACIONES COMPLEMENTARIAS
# =============================================================================

print("\n6. GENERANDO VISUALIZACIONES COMPLEMENTARIAS")
print("-" * 50)

# Crear figura con múltiples subplots
fig, axes = plt.subplots(2, 2, figsize=(16, 12))
fig.suptitle('Análisis Comprehensivo: Puntajes de Inglés por Estrato Socioeconómico', 
             fontsize=16, fontweight='bold', y=0.98)

# Gráfico 1: Medianas con intervalos de confianza
positions = range(len(estratos))
medianas = [est['Mediana'] for est in estadisticos_por_estrato]
ic_inf = [est['IC95%_inf'] for est in estadisticos_por_estrato]
ic_sup = [est['IC95%_sup'] for est in estadisticos_por_estrato]

axes[0,0].errorbar(positions, medianas, 
                   yerr=[np.array(medianas) - np.array(ic_inf), 
                         np.array(ic_sup) - np.array(medianas)], 
                   fmt='o', capsize=5, capthick=2, linewidth=2, color='steelblue')
axes[0,0].set_xticks(positions)
axes[0,0].set_xticklabels([f'Estrato {int(e)}\n(n={est["n"]:,})' 
                          for e, est in zip(estratos, estadisticos_por_estrato)])
axes[0,0].set_ylabel('Puntaje de Inglés')
axes[0,0].set_title('Medianas con IC 95% por Estrato', fontweight='bold')
axes[0,0].grid(True, alpha=0.3)

# Gráfico 2: Heatmap de valores p de Dunn
mask = np.triu(np.ones_like(dunn_result, dtype=bool), k=1)
im = axes[0,1].imshow(dunn_result.values, cmap='RdYlBu_r', aspect='auto', vmin=0, vmax=0.1)
axes[0,1].set_xticks(range(len(estratos)))
axes[0,1].set_yticks(range(len(estratos)))
axes[0,1].set_xticklabels([f'E{int(e)}' for e in estratos])
axes[0,1].set_yticklabels([f'E{int(e)}' for e in estratos])
axes[0,1].set_title('Matriz de p-values Ajustados (Dunn)', fontweight='bold')

# Añadir valores al heatmap
for i in range(len(estratos)):
    for j in range(len(estratos)):
        if i != j:
            text = axes[0,1].text(j, i, f'{dunn_result.iloc[i, j]:.3f}',
                        ha="center", va="center", color="black", fontsize=8)

# Gráfico 3: Heatmap de tamaños del efecto Cliff's Delta
delta_matrix = pd.DataFrame(np.nan, index=dunn_result.index, columns=dunn_result.columns)
for _, row in df_effect_sizes.iterrows():
    i = int(row['Grupo_A'].split()[-1])
    j = int(row['Grupo_B'].split()[-1])
    delta_matrix.loc[f'Estrato {i}', f'Estrato {j}'] = row['Cliffs_Delta']
    delta_matrix.loc[f'Estrato {j}', f'Estrato {i}'] = -row['Cliffs_Delta']

mask = np.triu(np.ones_like(delta_matrix, dtype=bool), k=1)
im3 = axes[1,0].imshow(delta_matrix.values, cmap='RdBu_r', aspect='auto', vmin=-1, vmax=1)
axes[1,0].set_xticks(range(len(estratos)))
axes[1,0].set_yticks(range(len(estratos)))
axes[1,0].set_xticklabels([f'E{int(e)}' for e in estratos])
axes[1,0].set_yticklabels([f'E{int(e)}' for e in estratos])
axes[1,0].set_title("Tamaños del Efecto (Cliff's Delta)", fontweight='bold')

# Añadir valores al heatmap de Cliff's Delta
for i in range(len(estratos)):
    for j in range(len(estratos)):
        if i != j and not np.isnan(delta_matrix.iloc[i, j]):
            text = axes[1,0].text(j, i, f'{delta_matrix.iloc[i, j]:.2f}',
                        ha="center", va="center", color="black", fontsize=8)

# Gráfico 4: Distribución por estrato
sns.boxplot(data=dunn_df, x='estrato', y='puntaje', ax=axes[1,1])
axes[1,1].set_title('Distribución de Puntajes por Estrato', fontweight='bold')
axes[1,1].set_xlabel('Estrato Socioeconómico')
axes[1,1].set_ylabel('Puntaje de Inglés')
axes[1,1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()

# =============================================================================
# 7. CONCLUSIÓN ESTADÍSTICA MEJORADA
# =============================================================================

print("\n7. CONCLUSIÓN ESTADÍSTICA MEJORADA")
print("═" * 70)

print(f"✓ La prueba de Kruskal-Wallis reveló diferencias estadísticamente significativas ")
print(f"  entre los estratos socioeconómicos (H = {stat_kw:.4f}, p = {p_kw:.4f}).")

print(f"✓ El tamaño del efecto global (epsilon-squared = {epsilon_sq:.4f}) indica un {magnitud}.")

# Identificar comparaciones significativas para la conclusión
comparaciones_con_efecto = []
for comp in effect_sizes:
    if abs(comp['Cliffs_Delta']) >= 0.11:  # Al menos efecto pequeño
        comparaciones_con_efecto.append(comp)

if comparaciones_con_efecto:
    print(f"✓ Se identificaron {len(comparaciones_con_efecto)} comparaciones con efectos relevantes:")
    for comp in comparaciones_con_efecto[:5]:  # Mostrar máximo 5
        print(f"  - {comp['Grupo_A']} vs {comp['Grupo_B']}: δ = {comp['Cliffs_Delta']:.3f} ({comp['Interpretacion']})")

print(f"✓ Los tamaños del efecto por pares (Cliff's Delta) varían desde {df_effect_sizes['Cliffs_Delta'].min():.3f}")
print(f"  hasta {df_effect_sizes['Cliffs_Delta'].max():.3f}, indicando diferentes magnitudes de diferencia entre estratos.")

print(f"✓ La tabla resumen proporciona medias, medianas, IC95% y diferencias significativas ")
print(f"  que sustentan estadísticamente las conclusiones del estudio.")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# Configuración de estilos
plt.style.use('default')
sns.set_palette("husl")

# =============================================================================
# 1. TABLA DE INTERPRETACIÓN DE CLIFF'S DELTA
# =============================================================================

def crear_tabla_probabilidades():
    """Crea una tabla con la interpretación de Cliff's Delta en términos de probabilidad"""
    
    # Valores de delta comunes y sus interpretaciones
    deltas = [0.0, 0.05, 0.11, 0.15, 0.20, 0.28, 0.33, 0.40, 0.43, 0.50, 0.60, 0.75, 0.90]
    
    tabla = []
    for delta in deltas:
        # Calcular probabilidad: P(Y > X) = (1 + delta) / 2
        probabilidad = (1 + delta) / 2 * 100
        
        # Interpretación estándar
        if abs(delta) < 0.11:
            interpretacion = "Efecto pequeño"
            magnitud = "Trivial"
        elif abs(delta) < 0.28:
            interpretacion = "Efecto medio"
            magnitud = "Pequeño"
        elif abs(delta) < 0.43:
            interpretacion = "Efecto grande"
            magnitud = "Moderado"
        else:
            interpretacion = "Efecto muy grande"
            magnitud = "Grande"
        
        # Descripción en lenguaje natural
        if delta == 0:
            descripcion = "Igual probabilidad (50%-50%)"
        elif delta > 0:
            descripcion = f"Grupo Y tiene {probabilidad:.1f}% de probabilidad de ser mayor"
        else:
            descripcion = f"Grupo X tiene {100-probabilidad:.1f}% de probabilidad de ser mayor"
        
        tabla.append({
            'Cliff_Delta (δ)': f'{delta:.2f}',
            'Probabilidad P(Y>X)': f'{probabilidad:.1f}%',
            'Interpretación': interpretacion,
            'Magnitud': magnitud,
            'Descripción': descripcion
        })
    
    return pd.DataFrame(tabla)

# Crear y mostrar la tabla
print("═" * 100)
print("TABLA DE INTERPRETACIÓN: CLIFF'S DELTA Y PROBABILIDADES")
print("═" * 100)

df_tabla = crear_tabla_probabilidades()
print(df_tabla.to_string(index=False))

# =============================================================================
# 2. GRÁFICO DE RELACIÓN CLIFF'S DELTA vs PROBABILIDAD
# =============================================================================

def crear_grafico_probabilidad():
    """Crea un gráfico que muestra la relación entre Cliff's Delta y la probabilidad"""
    
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))
    
    # Rango completo de valores de delta
    delta_range = np.linspace(-1, 1, 500)
    probabilidad_range = (1 + delta_range) / 2 * 100
    
    # Gráfico 1: Relación matemática completa
    ax1.plot(delta_range, probabilidad_range, linewidth=3, color='steelblue')
    ax1.set_xlabel("Cliff's Delta (δ)", fontsize=12, fontweight='bold')
    ax1.set_ylabel('Probabilidad P(Y > X) (%)', fontsize=12, fontweight='bold')
    ax1.set_title('Relación: Cliff\'s Delta vs Probabilidad', fontsize=14, fontweight='bold')
    ax1.grid(True, alpha=0.3)
    
    # Añadir líneas de referencia para los puntos de corte
    puntos_corte = [-0.43, -0.28, -0.11, 0, 0.11, 0.28, 0.43]
    colores = ['red', 'orange', 'yellow', 'green', 'yellow', 'orange', 'red']
    
    for i, corte in enumerate(puntos_corte):
        prob_corte = (1 + corte) / 2 * 100
        ax1.axvline(x=corte, color=colores[i], linestyle='--', alpha=0.7)
        ax1.axhline(y=prob_corte, color=colores[i], linestyle='--', alpha=0.3)
        ax1.annotate(f'δ={corte:.2f}\n({prob_corte:.1f}%)', 
                    xy=(corte, prob_corte), 
                    xytext=(10, 10), textcoords='offset points',
                    bbox=dict(boxstyle="round,pad=0.3", facecolor='white', alpha=0.8),
                    fontsize=8)
    
    # Gráfico 2: Zonas de interpretación
    ax2.set_xlim(-1, 1)
    ax2.set_ylim(0, 100)
    
    # Definir zonas de efecto
    zonas = [
        (-1.0, -0.43, 'Efecto muy grande\n(Grupo X domina)', 'lightcoral'),
        (-0.43, -0.28, 'Efecto grande\n(Grupo X domina)', 'lightpink'),
        (-0.28, -0.11, 'Efecto medio\n(Grupo X domina)', 'lightyellow'),
        (-0.11, 0.11, 'Efecto pequeño\n(Sin dominancia clara)', 'lightgreen'),
        (0.11, 0.28, 'Efecto medio\n(Grupo Y domina)', 'lightyellow'),
        (0.28, 0.43, 'Efecto grande\n(Grupo Y domina)', 'lightpink'),
        (0.43, 1.0, 'Efecto muy grande\n(Grupo Y domina)', 'lightcoral')
    ]
    
    for zona in zonas:
        inicio, fin, texto, color = zona
        prob_inicio = (1 + inicio) / 2 * 100
        prob_fin = (1 + fin) / 2 * 100
        
        ax2.axvspan(inicio, fin, alpha=0.3, color=color)
        ax2.text((inicio + fin) / 2, 50, texto, 
                ha='center', va='center', fontweight='bold',
                bbox=dict(boxstyle="round,pad=0.5", facecolor='white', alpha=0.8))
    
    ax2.set_xlabel("Cliff's Delta (δ)", fontsize=12, fontweight='bold')
    ax2.set_ylabel('Probabilidad P(Y > X) (%)', fontsize=12, fontweight='bold')
    ax2.set_title('Zonas de Interpretación de Cliff\'s Delta', fontsize=14, fontweight='bold')
    ax2.grid(True, alpha=0.3)
    
    # Añadir escala de probabilidad en el eje Y derecho
    ax2_twin = ax2.twinx()
    ax2_twin.set_ylim(0, 100)
    ax2_twin.set_ylabel('Probabilidad P(X > Y) (%)', fontsize=12, fontweight='bold')
    
    plt.tight_layout()
    plt.show()
    
    return fig

print("\n" + "═" * 100)
print("GRÁFICO DE INTERPRETACIÓN VISUAL")
print("═" * 100)

fig = crear_grafico_probabilidad()

# =============================================================================
# 3. CALCULADORA INTERACTIVA PARA TUS DATOS
# =============================================================================

def calcular_probabilidades_tus_datos(df_effect_sizes):
    """Calcula las probabilidades específicas para tus resultados"""
    
    print("\n" + "═" * 100)
    print("PROBABILIDADES ESPECÍFICAS PARA TUS COMPARACIONES")
    print("═" * 100)
    
    resultados = []
    
    for _, fila in df_effect_sizes.iterrows():
        delta = fila['Cliffs_Delta']
        probabilidad = (1 + delta) / 2 * 100
        
        # Determinar qué grupo domina
        if delta > 0:
            grupo_dominante = fila['Grupo_A']
            grupo_receptor = fila['Grupo_B']
            descripcion = f"{grupo_dominante} domina sobre {grupo_receptor}"
        else:
            grupo_dominante = fila['Grupo_B'] 
            grupo_receptor = fila['Grupo_A']
            descripcion = f"{grupo_dominante} domina sobre {grupo_receptor}"
            probabilidad = 100 - probabilidad  # Invertir para el grupo dominante
        
        resultados.append({
            'Comparación': f"{fila['Grupo_A']} vs {fila['Grupo_B']}",
            'Cliff_Delta': f'{delta:.4f}',
            'Probabilidad_Dominancia': f'{probabilidad:.1f}%',
            'Grupo_Dominante': grupo_dominante,
            'Interpretación': fila['Interpretacion'],
            'Descripción': descripcion
        })
    
    df_resultados = pd.DataFrame(resultados)
    print(df_resultados.to_string(index=False))
    
    return df_resultados

# =============================================================================
# EJECUCIÓN PRINCIPAL
# =============================================================================

if __name__ == "__main__":
    # 1. Mostrar tabla general de interpretación
    df_tabla = crear_tabla_probabilidades()
    
    # 2. Mostrar gráficos generales
    fig = crear_grafico_probabilidad()
      
    print("\n" + "═" * 100)
    print("RESUMEN DE INTERPRETACIÓN")
    print("═" * 100)
    print("""
    CÓMO INTERPRETAR CLIFF'S DELTA EN TÉRMINOS DE PROBABILIDAD:
    
    • Cliff's Delta (δ) = -1.0 → 0% probabilidad de que Y > X (X siempre domina)
    • Cliff's Delta (δ) =  0.0 → 50% probabilidad (igualdad perfecta)  
    • Cliff's Delta (δ) = +1.0 → 100% probabilidad de que Y > X (Y siempre domina)
    
    FÓRMULA: Probabilidad P(Y > X) = (1 + δ) / 2 * 100%
    
    En tu contexto educativo:
    - δ = 0.20 significa que hay 60% de probabilidad de que un estudiante del 
      estrato más alto tenga mejor puntaje que uno del estrato más bajo
    - δ = -0.30 significa que hay 35% de probabilidad (65% en sentido contrario)
    """)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import statsmodels.api as sm
from statsmodels.stats.power import TTestPower

# Configuración de estilo para gráficos
plt.style.use('default')
sns.set_palette("husl")

# Cargar los datos
df = pd.read_excel(r"D:\Anaconda\Saber Pro 20232.xlsx", sheet_name='DATA_ICFES_SABERPRO_PROFESIONAL')

# Definir los nombres de columnas
modalidad_col = 'estu_metodo_prgm'
puntaje_col = 'punt_global'
facultad_col = 'gruporeferencia'

# Filtrar solo estudiantes de ingeniería
df_ingenieria = df[df[facultad_col].str.contains('INGENIERIA|INGENIERÍA', case=False, na=False)]

# Filtrar solo PRESENCIAL y DISTANCIA
modalidades_validas = ['PRESENCIAL', 'DISTANCIA']
df_ingenieria = df_ingenieria[df_ingenieria[modalidad_col].isin(modalidades_validas)]

# Separar los grupos
presencial = df_ingenieria[df_ingenieria[modalidad_col] == 'PRESENCIAL'][puntaje_col]
distancia = df_ingenieria[df_ingenieria[modalidad_col] == 'DISTANCIA'][puntaje_col]

# Limpiar valores nulos
presencial = presencial.dropna()
distancia = distancia.dropna()

print(f"Tamaño de muestras:")
print(f"Presencial: {len(presencial)} estudiantes")
print(f"Distancia: {len(distancia)} estudiantes")

# Estadísticas descriptivas
print(f"\nESTADÍSTICAS DESCRIPTIVAS:")
print(f"Presencial - Media: {presencial.mean():.2f}, DE: {presencial.std():.2f}, Varianza: {presencial.var():.2f}")
print(f"Distancia - Media: {distancia.mean():.2f}, DE: {distancia.std():.2f}, Varianza: {distancia.var():.2f}")

# PRUEBA DE HOMOGENEIDAD DE VARIANZAS (LEVENE)
print(f"\n{'='*80}")
print("PRUEBA DE HOMOGENEIDAD DE VARIANZAS - LEVENE")
print(f"{'='*80}")

# Realizar prueba de Levene
levene_stat, levene_p = stats.levene(presencial, distancia)

print(f"Estadístico de Levene: {levene_stat:.4f}")
print(f"Valor p de Levene: {levene_p:.6f}")
print(f"Grados de libertad: {1} (entre grupos), {len(presencial) + len(distancia) - 2} (dentro de grupos)")

# Interpretación de la prueba de Levene
alpha_levene = 0.05
print(f"\nHipótesis de la prueba de Levene:")
print(f"H0: Las varianzas de ambos grupos son iguales (homocedasticidad)")
print(f"H1: Las varianzas de ambos grupos son diferentes (heterocedasticidad)")

if levene_p > alpha_levene:
    print(f"✓ p-value ({levene_p:.6f}) > α ({alpha_levene})")
    print(f"✓ No se rechaza H0: Las varianzas son homogéneas")
    print(f"✓ Se puede usar la prueba T con varianzas iguales (equal_var=True)")
    equal_var_flag = True
else:
    print(f"✗ p-value ({levene_p:.6f}) ≤ α ({alpha_levene})")
    print(f"✗ Se rechaza H0: Las varianzas NO son homogéneas")
    print(f"✗ Se recomienda usar la prueba T con varianzas diferentes (equal_var=False)")
    equal_var_flag = False

# Razón de varianzas (F-test)
var_ratio = max(presencial.var(), distancia.var()) / min(presencial.var(), distancia.var())
print(f"\nRazón de varianzas (mayor/menor): {var_ratio:.2f}")

# Regla empírica para homogeneidad de varianzas
if var_ratio < 4:
    print(f"✓ La razón de varianzas es menor que 4: las varianzas se consideran suficientemente homogéneas")
else:
    print(f"⚠️  La razón de varianzas es mayor que 4: puede haber problemas de heterocedasticidad")

# Cálculo manual de la prueba T para mejor comprensión
print(f"\n{'='*80}")
print("CÁLCULO DETALLADO DE LA PRUEBA T STUDENT (1 COLA)")
print(f"{'='*80}")

# 1. Cálculo del estadístico t manualmente
n1, n2 = len(presencial), len(distancia)
mean1, mean2 = presencial.mean(), distancia.mean()
var1, var2 = presencial.var(ddof=1), distancia.var(ddof=1)  # varianza muestral

# Varianza agrupada (pooled variance) - solo para equal_var=True
pooled_var = ((n1 - 1) * var1 + (n2 - 1) * var2) / (n1 + n2 - 2)
pooled_std = np.sqrt(pooled_var)

# Error estándar de la diferencia (diferente según equal_var)
if equal_var_flag:
    se_diff = pooled_std * np.sqrt(1/n1 + 1/n2)
    method = "varianzas iguales"
else:
    # Welch's t-test (varianzas diferentes)
    se_diff = np.sqrt(var1/n1 + var2/n2)
    method = "varianzas diferentes (Welch)"

# Estadístico t calculado manualmente
t_calculado = (mean1 - mean2) / se_diff

print(f"1. CÁLCULO DEL ESTADÍSTICO t ({method}):")
print(f"   Media presencial (X̄1): {mean1:.4f}")
print(f"   Media distancia (X̄2): {mean2:.4f}")
print(f"   Diferencia de medias (X̄1 - X̄2): {mean1 - mean2:.4f}")
if equal_var_flag:
    print(f"   Varianza agrupada (s²_p): {pooled_var:.4f}")
    print(f"   Desviación estándar agrupada (s_p): {pooled_std:.4f}")
print(f"   Error estándar de la diferencia: {se_diff:.4f}")
print(f"   Estadístico t CALCULADO: {t_calculado:.4f}")

# 2. Grados de libertad
if equal_var_flag:
    df_degrees = n1 + n2 - 2
    print(f"\n2. GRADOS DE LIBERTAD (varianzas iguales):")
    print(f"   gl = n1 + n2 - 2 = {n1} + {n2} - 2 = {df_degrees}")
else:
    # Grados de libertad para Welch's t-test
    df_degrees = ((var1/n1 + var2/n2)**2) / ((var1/n1)**2/(n1-1) + (var2/n2)**2/(n2-1))
    print(f"\n2. GRADOS DE LIBERTAD (Welch-Satterthwaite):")
    print(f"   gl ≈ {df_degrees:.1f}")

# 3. Valor crítico de t (1 cola, α = 0.05)
alpha = 0.05
t_critical = stats.t.ppf(1 - alpha, df_degrees)
print(f"\n3. VALOR CRÍTICO DE t:")
print(f"   Nivel de significancia (α): {alpha}")
print(f"   Prueba de una cola (cola derecha)")
print(f"   Grados de libertad: {df_degrees:.1f}")
print(f"   Valor crítico de t: {t_critical:.4f}")

# 4. Valor p
p_value = 1 - stats.t.cdf(t_calculado, df_degrees)  # Cálculo manual del p-value
print(f"\n4. VALOR p:")
print(f"   Valor p calculado: {p_value:.6f}")

# 5. Región de rechazo y decisión
print(f"\n5. DECISIÓN ESTADÍSTICA:")
print(f"   Hipótesis nula (H0): μ_presencial ≤ μ_distancia")
print(f"   Hipótesis alternativa (H1): μ_presencial > μ_distancia")
print(f"   Estadístico t calculado: {t_calculado:.4f}")
print(f"   Valor crítico de t: {t_critical:.4f}")

if t_calculado > t_critical:
    print(f"   ✓ t_calculado ({t_calculado:.4f}) > t_crítico ({t_critical:.4f})")
    print(f"   ✓ Se rechaza la hipótesis nula")
else:
    print(f"   ✗ t_calculado ({t_calculado:.4f}) ≤ t_crítico ({t_critical:.4f})")
    print(f"   ✗ No se rechaza la hipótesis nula")

print(f"\n   Valor p: {p_value:.6f}")
if p_value < alpha:
    print(f"   ✓ p-value ({p_value:.6f}) < α ({alpha})")
    print(f"   ✓ Se rechaza la hipótesis nula")
else:
    print(f"   ✗ p-value ({p_value:.6f}) ≥ α ({alpha})")
    print(f"   ✗ No se rechaza la hipótesis nula")

# 6. También usando scipy para verificación
t_stat_scipy, p_value_scipy = stats.ttest_ind(presencial, distancia, equal_var=equal_var_flag, alternative='greater')
print(f"\n6. VERIFICACIÓN CON SCIPY ({method}):")
print(f"   Estadístico t (scipy): {t_stat_scipy:.4f}")
print(f"   Valor p (scipy): {p_value_scipy:.6f}")

# 7. Intervalo de confianza para la diferencia de medias
conf_int = stats.t.interval(confidence=0.95, 
                           df=df_degrees,
                           loc=mean1 - mean2,
                           scale=se_diff)

print(f"\n7. INTERVALO DE CONFIANZA 95%:")
print(f"   Diferencia de medias: {mean1 - mean2:.4f}")
print(f"   Límite inferior: {conf_int[0]:.4f}")
print(f"   Límite superior: {conf_int[1]:.4f}")

# 8. Tamaño del efecto (Cohen's d)
if equal_var_flag:
    pooled_std_effect = pooled_std
else:
    # Para varianzas diferentes, usar desviación estándar promedio ponderada
    pooled_std_effect = np.sqrt((var1 + var2) / 2)

cohens_d = (mean1 - mean2) / pooled_std_effect
print(f"\n8. TAMAÑO DEL EFECTO:")
print(f"   Cohen's d: {cohens_d:.4f}")

# Interpretación del tamaño del efecto
if abs(cohens_d) < 0.2:
    magnitud = "Muy pequeño"
elif abs(cohens_d) < 0.5:
    magnitud = "Pequeño"
elif abs(cohens_d) < 0.8:
    magnitud = "Mediano"
else:
    magnitud = "Grande"
print(f"   Magnitud: {magnitud}")

# 9. Potencia estadística
print(f"\n9. POTENCIA ESTADÍSTICA:")
try:
    power_analysis = TTestPower()
    power = power_analysis.solve_power(
        effect_size=cohens_d,
        nobs1=n1,
        alpha=alpha,
        power=None,
        alternative='larger'
    )
    print(f"   Potencia (1-β): {power:.4f}")
    if power < 0.8:
        print(f"   ⚠️  Potencia insuficiente (< 0.8)")
    else:
        print(f"   ✓ Potencia adecuada (≥ 0.8)")
except:
    print(f"   No se pudo calcular la potencia")

print(f"\n{'='*80}")
print("RESUMEN EJECUTIVO")
print(f"{'='*80}")
print(f"Prueba de Levene: p = {levene_p:.6f} {'(Homogéneas)' if levene_p > 0.05 else '(Heterogéneas)'}")
print(f"Método T-test usado: {method}")
print(f"Estadístico t: {t_calculado:.4f}")
print(f"Valor crítico de t (α=0.05, 1 cola): {t_critical:.4f}")
print(f"Valor p: {p_value:.6f}")
print(f"Decisión: {'Rechazar H0' if p_value < alpha else 'No rechazar H0'}")
print(f"Diferencia de medias: {mean1 - mean2:.4f}")
print(f"Intervalo de confianza 95%: ({conf_int[0]:.4f}, {conf_int[1]:.4f})")
print(f"Tamaño del efecto (Cohen's d): {cohens_d:.4f} ({magnitud})")

# Gráfico de la distribución t y región crítica
plt.figure(figsize=(12, 6))
x = np.linspace(-4, 4, 1000)
y = stats.t.pdf(x, df_degrees)

plt.plot(x, y, 'b-', label=f'Distribución t (gl={df_degrees:.1f})')
plt.axvline(x=t_critical, color='r', linestyle='--', 
           label=f'Valor crítico t = {t_critical:.3f}')
plt.axvline(x=t_calculado, color='g', linestyle='-', 
           label=f'Estadístico t = {t_calculado:.3f}')

# Sombrear región de rechazo
x_reject = np.linspace(t_critical, 4, 100)
y_reject = stats.t.pdf(x_reject, df_degrees)
plt.fill_between(x_reject, y_reject, color='red', alpha=0.3, label='Región de rechazo (α=0.05)')

plt.title(f'Distribución t Student y Región Crítica (1 Cola)\n{method}')
plt.xlabel('Valor t')
plt.ylabel('Densidad de probabilidad')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

# Gráfico comparativo de medias con intervalo de confianza
plt.figure(figsize=(10, 6))
categorias = ['Presencial', 'Distancia', 'Diferencia']
valores = [mean1, mean2, mean1 - mean2]
errores = [presencial.std()/np.sqrt(n1), distancia.std()/np.sqrt(n2), se_diff]

colors = ['skyblue', 'lightcoral', 'lightgreen']
bars = plt.bar(categorias[:2], valores[:2], yerr=errores[:2], capsize=10, 
              alpha=0.7, color=colors[:2])

plt.title('Comparación de Puntajes por Modalidad\n(Con intervalos de confianza 95%)')
plt.ylabel('Puntaje Global')
plt.grid(True, alpha=0.3)

# Añadir valores en las barras
for bar, valor in zip(bars, valores[:2]):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 5, 
            f'{valor:.1f}', ha='center', va='bottom', fontweight='bold')

plt.tight_layout()
plt.show()

# GRÁFICA DE DENSIDAD DETALLADA
plt.figure(figsize=(14, 8))

# Gráfico de densidad principal
sns.kdeplot(presencial, label='Presencial', fill=True, alpha=0.6, color='blue', linewidth=2)
sns.kdeplot(distancia, label='Distancia', fill=True, alpha=0.6, color='red', linewidth=2)

# Líneas verticales para las medias
plt.axvline(mean1, color='blue', linestyle='--', linewidth=2, alpha=0.8, 
           label=f'Media Presencial: {mean1:.2f}')
plt.axvline(mean2, color='red', linestyle='--', linewidth=2, alpha=0.8,
           label=f'Media Distancia: {mean2:.2f}')

# Sombrear la diferencia entre medias
x_fill = np.linspace(mean2, mean1, 100)
y1_fill = stats.gaussian_kde(presencial)(x_fill)
y2_fill = stats.gaussian_kde(distancia)(x_fill)
plt.fill_between(x_fill, y1_fill, y2_fill, color='purple', alpha=0.3, 
                label='Área de diferencia')

# Configuración del gráfico
plt.title('Distribución de Densidad de Puntajes Saber Pro\nModalidad Presencial vs Distancia en Ingeniería', 
          fontsize=16, fontweight='bold', pad=20)
plt.xlabel('Puntaje Global Saber Pro', fontsize=12, fontweight='bold')
plt.ylabel('Densidad de Probabilidad', fontsize=12, fontweight='bold')
plt.legend(fontsize=11, loc='upper right')

# Añadir información estadística en el gráfico
textstr = '\n'.join([
    f'Muestra Presencial: n = {n1}',
    f'Muestra Distancia: n = {n2}',
    f'Media Presencial: {mean1:.2f}',
    f'Media Distancia: {mean2:.2f}',
    f'Diferencia: {mean1 - mean2:.2f}',
    f't = {t_calculado:.3f}',
    f'p-value = {p_value:.4f}',
    f'Cohen\'s d = {(mean1 - mean2)/pooled_std:.3f}'
])

props = dict(boxstyle='round', facecolor='wheat', alpha=0.8)
plt.text(0.02, 0.98, textstr, transform=plt.gca().transAxes, fontsize=10,
         verticalalignment='top', bbox=props)

# Añadir grid y mejorar estética
plt.grid(True, alpha=0.3, linestyle='--')
plt.xlim(min(presencial.min(), distancia.min()) - 10, 
         max(presencial.max(), distancia.max()) + 10)

# Añadir anotación de significancia
if p_value < alpha:
    significance_text = 'Diferencia estadísticamente significativa (p < 0.05)'
    significance_color = 'green'
else:
    significance_text = 'Diferencia no significativa (p ≥ 0.05)'
    significance_color = 'red'

plt.annotate(significance_text, 
            xy=(0.5, 0.95), xycoords='axes fraction',
            fontsize=12, fontweight='bold', color=significance_color,
            ha='center', va='top',
            bbox=dict(boxstyle="round,pad=0.3", facecolor="white", alpha=0.8))

plt.tight_layout()
plt.show()
```
